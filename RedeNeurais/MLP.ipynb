```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
X = pd.DataFrame(np.array([np.ones(13),np.arange(0,13)]).T)
y = np.array([150,130,114,94,78,62,55,44,41,33,27,18,17])
```

```python
w1 = np.random.normal(0,1,(3,2))

w2 = np.random.normal(0,1,(2,4))

w3 = np.random.normal(0,1,(1,3))

n = 0.005
```

```python
def I(X,w):
    return np.dot(X,w.T)
```

```python
def sigmoid(x, derivada = False):
    if not derivada:
      return 1/(1+np.exp(-x))
    else:
      return 1/(1+np.exp(-x))* (1-1/(1+np.exp(-x)))
```

```python
def relu(x, derivada = False):
  if not derivada:
    return np.maximum(0, x)

  else:
    return np.where(x > 0, 1, 0)
```

```python
def tanh(x, derivada = False):
  if not derivada:
    return np.tanh(x)
  else:
    return 1 - np.tanh(x)**2
```

```python
def linear(x,derivada = False):
  if not derivada:
    return x
  else:
    return np.ones_like(x)
```

```python
def L(y,y_pred):
  return 1/2*np.sum((y-y_pred)**2)
```

```python
epoch = 1000
for i in range(epoch):
  ##forward propagation
  y1 = relu(I(X,w1))
  aux = np.append(np.ones((13,1)),y1,axis = 1)

  y2 = relu(I(aux,w2))
  aux2 = np.append(np.ones((13,1)),y2,axis = 1)

  y3 = relu(I(aux2,w3))
  delta_3 = (y - y3)@relu(I(aux2,w3),True)
  w3 = w3 + (n * np.dot(delta_3.T,aux2))

  ##backforward
  delta_2 = np.sum(delta_3@w3, axis =1,keepdims=True)*relu(I(aux,w2),True)
  w2 = w2 + (n * np.dot(delta_2.T,aux))

  delta_1 = np.sum(delta_2@w2, axis =1,keepdims=True)*relu(I(X,w1),True)
  w1 = w1 + (n * np.dot(delta_1.T,X))



```
