```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

x0 = np.ones(len(dados))*-1

x0_colum = pd.DataFrame(x0)
df_0 = pd.DataFrame(dados)

df = pd.concat([x0_colum,df_0],axis=1,ignore_index=True)


X = df.iloc[: , 0:4]
Y = df.iloc[: , -1]

X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size = 0.15,random_state=32)


```

###Definindo o potencial de ativação // Integração

```python
def u(X,w,i):
  u = np.dot(X.iloc[i,:].to_numpy(),w)

  return u
```

```python
#Definindo os pesos iniciais
w = np.array([0.2,0.2,0.2,0.2])

```

###Função de ativação

```python
def g(u):
  if u >= 0:
    return 1

  else:
    return -1
```

###Treinamento do percepton de camada única


```python
def trainingHabb(X,Y,w,n,):

  epoca = 0
  while True:
    erro = 0

    for i in range(len(X)):
      U = u(X,w,i)
      y = g(U)
      if Y.iloc[i] != y:
        w = w + n*(Y.iloc[i] - y)*X.iloc[i,:].to_numpy()
        erro = 1

    epoca = epoca + 1

    if erro == 0:
      return epoca,w


```

```python
n = 0.2
epoca,w_otimo = trainingHabb(X_train,y_train,w,n)

print(f'Vetor de pesos após o treinamento {w_otimo} \nEpoca necessárias {epoca}')

#Vale enfatiza que o treinamento foi realizado com todos os dados disponível
#Depois separar os dados em conjuto de treino e teste e ai sim testar o modelo
```

```python
ctd = 0
for i in range(len(X_test)):
  U = u(X_test,w,i)
  y = g(U)

  if y == y_test.iloc[i]:
    ctd = ctd + 1


#Acurácia

acc = (ctd/len(X_test))*100

print(f'Acurácia do modelo {acc}%')
```
