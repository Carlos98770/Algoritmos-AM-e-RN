#Algoritmo KNN


```python
def knn(df, queryPoint, K):
	a = np.sum(((queryPoint - df.iloc[: , 0:len(df.columns) -1])**2),axis=1)**0.5
	ids = a.sort_values().index
	classesK = []


	for i in range(K):
		classesK.append(df.iloc[ids[i],-1])


	y = pd.Series(classesK)
	valores = y.value_counts()
	classification = valores.idxmax()

	return classification
```

```python

import numpy as np
import pandas as pd


def knn(df, queryPoint, K):
	a = np.sum(((queryPoint - df.iloc[: , 0:len(df.columns) -1])**2),axis=1)**0.5
	ids = a.sort_values().index
	classesK = []


	for i in range(K):
		classesK.append(df.iloc[ids[i],-1])


	y = pd.Series(classesK)
	valores = y.value_counts()
	classification = valores.idxmax()

	return classification,valores

def main():
	dados = {
		"Temperatura" : [34,32,26,22],
		"Umidade" : [78,65,89,56],
		"playOrnot" : ['N','N','S','S']

	}

	df = pd.DataFrame(dados)
	queryPoint= np.array([32,79])
	K = 3

	classification,votos = knn(df, queryPoint, K)

	print(f"Classificação para classe: {classification} \nVotos: {votos}")


if __name__ == "__main__":
	main()
```
