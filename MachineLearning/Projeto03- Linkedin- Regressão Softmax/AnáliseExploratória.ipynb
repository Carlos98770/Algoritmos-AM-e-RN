```python
df['Temperature (K)'] = df['Temperature (K)'].astype(float)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(range(len(df['Temperature (K)'])), df['Temperature (K)'], color='purple', alpha=0.6)
plt.title('Gráfico de Dispersão dos Dados')
plt.xlabel('Índice')
plt.ylabel('Valor')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.boxplot(df['Temperature (K)'], patch_artist=True, vert=False,
            boxprops=dict(facecolor='lightgreen'))

plt.title('Boxplot dos Dados')
plt.xlabel('Valor')
plt.show()
```
