###Regressão logistíca, Algoritmo de classificação binária

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
```

```python
X,y = load_iris(return_X_y=True, as_frame=True)

X_logist = X.iloc[:,3:]

y_logist = np.zeros(len(y))

for i in range(len(y)):
  if y[i] == 2:
    y_logist[i] = 0

  else:
    y_logist[i] = y[i]

X_logist
```

```python
log_reg = LogisticRegression()
log_reg.fit(X_logist,y_logist)

y_pred = log_reg.predict_proba(X_logist)

y_pred
```

```python
plt.plot(X_logist,y_pred[:,0],"g-",label = 'Iris-Virginica')
plt.plot(X_logist,y_pred[:,1],"b--",label = 'Not Iris-Virginica')

plt.show()
```

```python
X_new = np.linspace(0,3,1000).reshape(-1,1)

y_prob = log_reg.predict_proba(X_new)
plt.plot(X_new,y_prob[:,1], 'g-')
plt.plot(X_new,y_prob[:,0] , 'b--')
plt.scatter(X_logist,y_logist)
```
