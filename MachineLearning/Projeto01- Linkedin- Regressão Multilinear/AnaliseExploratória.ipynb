#Tratamento dos dados para regressão

* Atributos de localização como a latitude e longitude será tratado com a distancia da capital até certo local

* Colunas region e state será removida, pois não terá nenhuma influencia na regressão




```python
estados_copy = estados.copy()
df_copy = df.copy()
for estado in list(estados_copy.keys()):
  df_copy = estados_copy[estado]
  df_copy.drop(['state','region'],axis=1,inplace = True)
  estados_copy[estado] = df_copy
```

```python
matriz_keys = list(estados_copy.keys())
matriz_keys
```

```python
cords_capitais = [                 #longitude          #latitude
                  [matriz_keys[0],-34.89066530710511,-8.047560298729962],
                  [matriz_keys[1], -35.212800502148355,-5.800530617556163],
                  [matriz_keys[2],-51.23349948060968,-30.029274183619485],
                  [matriz_keys[3], -43.220630529833954,-22.9103982941391],
                  [matriz_keys[4],-48.5437631376258,-27.57664641267408],
                  #[matriz_keys[5],-37.08788025450991,-10.893889023744052],
                  [matriz_keys[5],-46.65132762385994,-23.54258092078885]

                  ]

cords_capitais
```

* Fazer a Distancia entre os pontos para criar uma nova coluna, (Distância da capital)



```python
import math

def distance(lat1, lon1, lat2, lon2):
    # Raio da Terra em quilômetros
    R = 6371.0

    # Converter graus para radianos
    lat1_rad = math.radians(lat1)
    lon1_rad = math.radians(lon1)
    lat2_rad = math.radians(lat2)
    lon2_rad = math.radians(lon2)

    # Diferenças
    dlat = lat2_rad - lat1_rad
    dlon = lon2_rad - lon1_rad

    # Fórmula de Haversine
    a = math.sin(dlat / 2)**2 + math.cos(lat1_rad) * math.cos(lat2_rad) * math.sin(dlon / 2)**2
    c = 2 * math.asin(math.sqrt(a))

    # Distância em quilômetros
    distance = R * c
    return distance
```

```python

for i in range(len(cords_capitais)):
  estado = estados_copy[cords_capitais[i][0]]
  lat1 = cords_capitais[i][2]
  lon1 = cords_capitais[i][1]

  distancias = []

  for j in range(len(estado)):
    lat2 = estado.iloc[j]['lat']
    lon2 = estado.iloc[j]['lon']
    distancias.append(distance(lat1,lon1,lat2,lon2))

  estado["Distância_a_capital(Km)"] = distancias
  estados_copy[cords_capitais[i][0]] = estado
```

* Estratégia a ser usada é fazer a media do estado

```python
for estado in list(estados_copy.keys()):
  imputer = SimpleImputer(missing_values=np.nan, strategy="median")
  imputer = imputer.fit(estados_copy[estado]['area_m2'].to_numpy().reshape(-1,1))
  estados_copy[estado]['area_m2'] = imputer.transform(estados_copy[estado]['area_m2'].to_numpy().reshape(-1,1))
```
