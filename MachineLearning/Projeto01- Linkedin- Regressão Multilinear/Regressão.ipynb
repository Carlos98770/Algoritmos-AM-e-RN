```python
y = df_copy.iloc[:,8]
df_aux = df_copy.drop('8',axis=1)
X = df_aux

```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.4,random_state=42)
```

#Aplicando a Regress√£o Multilinear


```python
class LinearRegression():

    #X e y devem ser um matriz 2d
    def __init__(self,X,y) -> None:
        self.__theta = np.random.rand(len(X.columns) + 1,1)
        self.__X = pd.DataFrame(X)
        self.__X.insert(0,"Bias",np.ones((len(X),)))
        self.__y = y.to_numpy().reshape(-1,1)

    def fit(self,learning_reating,n_interations) -> None:
        m = len(self.__X)
        for i in range(n_interations):
            grad = 2/m * self.__X.T @ (self.__X@self.__theta - self.__y)
            self.__theta = self.__theta - learning_reating*grad


    def predict(self,data) -> np.float32:
        data.insert(0,"Bias",np.ones((len(data),)))
        predict = data@self.__theta
        return predict

    def getTheta(self) -> np.array:
        return self.__theta

    def plot(self) -> np.array:
        return self.__X@self.__theta
```

```python
obj = LinearRegression(X_train,y_train)
```

```python
obj.fit(0.2,100)
```

```python
y_pred = obj.predict(X_test)
```

```python
mse = mean_squared_error(y_test,y_pred)
mae = mean_absolute_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)

print(f'MSE : {mse}')
print(f'MAE : {mae}')
print(f'R2 score: {r2}')
```

Isso mostra que os dados n√£o condiz com um ajuste linear, pois a metrica R2Score est√° bem baixa, isso implica que o modelo n√£o explica bem a varia√ß√£o dos dados por n motivos.
